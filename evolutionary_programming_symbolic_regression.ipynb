## EVOLUTIONARY PROGRAMMING SYMBOLIC REGRESSION

### PROBLEM
fit a regression line to a data set. the data set can be found in the benchmark problems of SRBench. This will be achieved through an algorithm implementing genetic programming. This will create candidate soluions in a tree structure, of functions in the nodes, and integers in the leaves. The population will be graded by how fitting the regression line is.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import treelib
```

#### HYPER PARAMETERS

```python
FUNC_LIST = [add,subtract,multiply,divide]
MAX_INT_FOR_LEAVES = 100
MIN_INT_FOR_LEAVES = -100
POPULATION_SIZE = 10
INITIAL_HEIGHT_SETTING = 6
X_PROBABILITY = 0.1
```

```python

def add(i,j):
    return i + j
def subtract(i,j):
    return i - j
def multiply(i,j):
    return i * j
def divide(i,j):
    return i / j

def generate_random_binary_tree(height):
    # generate root
    tree = treelib.Tree()
    root_node_value = np.random.choice(FUNC_LIST)
    root_node = treelib.node.Node(data=root_node_value)
    tree.add_node(root_node)

    #add nodes to tree
    #stops when a node reaches height 
    while tree.depth() < height:
        new_node_value = np.random.choice(FUNC_LIST)
        list_of_posible_parents = tree.leaves()
        
        for i in tree.all_nodes():
            if len(tree.children(nid=i.identifier)) == 1:
                list_of_posible_parents.append(i)

        new_node_parent = np.random.choice(list_of_posible_parents)
        new_node = treelib.node.Node(data=new_node_value)
        tree.add_node(parent=new_node_parent, node=new_node)

    #make sure that the tree is balanced
    needs_a_leaf = []
    for j in tree.all_nodes():
        if len(tree.children(nid=j.identifier)) == 1:
            needs_a_leaf.append(j)
    while needs_a_leaf != []:
        parent = needs_a_leaf.pop()
        value = np.random.randint(MAX_INT_FOR_LEAVES)
        new_node = treelib.node.Node(data=value)
        tree.add_node(parent=parent, node=new_node)

    for l in tree.leaves():
        random_chance_of_X = X_PROBABILITY
        if random_chance_of_X > np.random.rand(1):
            tree[l.identifier].data ="X"
            
        else:
            tree[l.identifier].data = np.random.randint(low = MIN_INT_FOR_LEAVES,high = MAX_INT_FOR_LEAVES)

    return tree

def create_population(pop_size):
    pop = []
    for i in range(pop_size):
        pop.append(generate_random_binary_tree(height=INITIAL_HEIGHT_SETTING))
    return pop

def tree_to_regression_line(tree):
    for nid in tree.expand_tree(mode=tree.DEPTH):
        print(tree[nid].data)
    #TO FINISH
    #
    #TO FINISH

def tree_to_num(tree, nid, input=1):
    if  tree[nid].is_leaf():
        print("tree[nid].is_leaf()", tree[nid].is_leaf())
        return tree[nid].data
    else:
        left_id = tree.children(nid)[0].identifier
        right_id = tree.children(nid)[1].identifier
        return tree[nid].data(tree_to_num(tree,left_id),tree_to_num(tree,right_id))
    
# def

    # for i in tree
```
